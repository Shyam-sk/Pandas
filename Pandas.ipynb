Filtering Data

```python
bios.tail()
```

```python
bios.info()
```

```python
from os import name
#filter players with height >215
bios[bios["height_cm"]>215]   # single condition
bios.loc[bios["height_cm"]>215,['name']]  # return only name
bios.loc[(bios["height_cm"]>215) & (bios["born_country"]=="USA")],['name'] # using '&' AND for multiple condition

# filter by string
bios[bios['name'].str.contains('keith', case=False)]  # filter by name contains "keith"
bios[bios['name'].str.contains('keith|sen', case=False)]  # filter by name contains "keith" OR "sen"

# GO through Regex

# isin
bios[bios["born_country"].isin(["USA","FRA","GBR"]) & (bios['name'].str.contains('keith|sen', case=False))]

# another methods

bios.query('born_country =="USA" and born_city == "Seattle"')
```

Adding/Removing Columsn

```python
# add new colum "Price" with value 49
# add new colum "new price with diff pirce for expresso and latte
# drop colmn "Price"

coffee["Price"] = 4.99
import numpy as np
coffee["New Price"] = np.where(coffee["Coffee Type"]=='Espresso', 3.99,5.99)
coffee
coffee.drop(columns=['Price'])  # this will drop column Price, but it will not modify df
# to drop column Price and modify table
coffee.drop(columns=['Price'], inplace=True)
coffee
#another method to remove price
# coffee = coffee[['Day','Coffee Type','Unit Sold','New Price]]


# rename new price to price
coffee.rename(columns={'New Price':'Price'}, inplace= True)
# add revenue column
coffee["Revenue"] = coffee['Units Sold'] * coffee['Price']
coffee
```
